```python
from PIL import Image
import cv2
import imageio

import numpy as np
from time import localtime, strftime

import requests
from io import BytesIO

import matplotlib.pyplot as plt
```

```python
frames = []

for i in range(30):
    response = requests.get("https://apps.lanecounty.org/LCWebCams/HarborVista")
    frames += [np.array( Image.open(BytesIO(response.content)).convert('RGB') )]

## for 30 frames, time = 1m 25.5s 
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(frames[29])
plt.show()
```

```python
imageio.mimsave('../harborvista_webcam/output.gif', frames, fps=0.75)
## time = (7.7s for 30 frames)
```

```python
print(f'../harborvista_webcam/{strftime("%m-%d_%H.%M", localtime())}_frames.gif')
```
